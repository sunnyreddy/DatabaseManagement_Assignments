# Create a physical schema

A physical data model is a representation of a data design as implemented, or intended to be implemented, in a database management system. In the lifecycle of a project it typically derives from a logical data model, though it may be reverse-engineered from a given database implementation.

1. Entities referred to as tables
2. Attributes referred to as columns
3. Database compatible table names

Questions you must answer about your physical model:
1. Are all the tables in 1NF?
2. Are all the tables in 2NF?
3. Are all the tables in 3NF?

First normal form (1NF)
* Each table has a primary key: minimal set of attributes which can uniquely identify a record
Example: players table has playerId as primary id which uniquie
* The values in each column of a table are atomic (No multi-value attributes allowed).
Example: height column contains height of the player (only single value)
* There are no repeating groups: two columns do not store similar information in the same table.
Every column in our database is unique and doesn't have any common data

Second normal form (2NF)
* All requirements for 1st NF met
* No partial dependencies
Example: None of player infomation is present in stats table which is in Players table (where you can directly access data using only playerid)
* No calculated data 
(because no column can be calculated from other)

Third normal form (3NF)
* All requirements for 2nd NF must be met.
* Eliminate fields that do not directly depend on the primary key; that is no transitive dependencies.
Example: Every table has attributes which are depend on that table

## PLAYER TABLE

```python
def nba_foxnews_scraper(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    table = soup.find_all("table",{"class": "wisbb_standardTable"})
    table_head = soup.find_all("thead",{"class": "wisbb_tableHeader"})
    return pd.read_html(str(table))[0]
```

```python
url2 = "https://www.foxsports.com/nba/players?teamId=0&season=2019&position=0&page=1&country=0&grouping=0&weightclass=0"
nba_foxnews_scraper(url2).head()
```

## We have so many pages of data so we pass those pages in url to get all data

```python
pages = list(i for i in range(1,28))

df_all_Playerdata_from_scraping = None

def get_all_players_data_scrapper(pages):
    i = 0
    for page in pages:
        url = "https://www.foxsports.com/nba/players?teamId=0&season=2019&position=0&page="+ str(page) + "&country=0&grouping=0&weightclass=0"        
        if i == 0:
            df_all_Playerdata_from_scraping = nba_foxnews_scraper(url)
            i = 1
        else:
            df_all_Playerdata_from_scraping = df_all_Playerdata_from_scraping.append(nba_foxnews_scraper(url))
    
    return df_all_Playerdata_from_scraping

df_all_Playerdata_from_scraping = get_all_players_data_scrapper(pages)
df_all_Playerdata_from_scraping.head()
```

```python
df_all_Playerdata_from_scraping.isnull().any()
```

## TEAMS

```python
url_wiki = "https://en.wikipedia.org/wiki/Wikipedia:WikiProject_National_Basketball_Association/National_Basketball_Association_team_abbreviations"
r = requests.get(url_wiki)
soup = BeautifulSoup(r.text, 'lxml')
table = soup.find_all("table")
df_teams = pd.read_html(str(table))[0]
df_teams.head()
```

```python
# deleting first row becoz not relavent
df_teams.drop(df_teams.head(1).index, inplace=True)
df_teams.columns = df_teams.columns.map(str)
df_teams = df_teams.rename(columns={"0": "abbrevation", "1": "teamName"})
df_teams
```

```python
df_teams.isnull().any()
```

## PLAYER STATS

```python
# Function to scrape all team names + respective http links from website
def get_teams(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    teams = soup.find_all("option")
    team_list = []
    i = 0
    for item in soup.find_all('option'):
        if i > 6:
            team_list.append(''.join(str(item['value'])))
        i += 1
    return team_list
```

```python
team_list = get_teams("https://www.foxsports.com/nba/boston-celtics-team-stats")
team_list
```

```python
def get_all_playerStats_data_scrapper(team_list):
    i = 0
    base_url = "https://www.foxsports.com"
    for team_url in team_list:
        url = base_url + team_url
        if i == 0:
            df_all_data_from_scraping = nba_foxnews_scraper(url)
            i = 1
        else:
            df_all_data_from_scraping = df_all_data_from_scraping.append(nba_foxnews_scraper(url))
    
    return df_all_data_from_scraping
        
df_all_data_from_scraping = get_all_playerStats_data_scrapper(team_list)
df_all_data_from_scraping.head()
```

```python
df_all_data_from_scraping.columns
```

```python
needed_columns_stats = ['SCORING','GP','GS','MPG','PPG','PPS','HIGH']
df_Player_Stats = df_all_data_from_scraping[needed_columns_stats]
df_Player_Stats.head()
```

```python
df_Player_Stats.isnull().any()
```
